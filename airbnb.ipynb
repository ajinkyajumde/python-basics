```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/drive/MyDrive/Almabetter projects/Airbnb EDA analysis/Airbnb NYC 2019.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.groupby('neighbourhood_group')['id'].count().reset_index(name='count')
```

```python
h=df.groupby(['room_type','neighbourhood_group'])['id'].count().unstack()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(25,10))
sns.lineplot(data=h, palette="tab10", linewidth=2.5)
```

```python
n=df.groupby(['neighbourhood_group'])['id'].count().reset_index(name='count').sort_values(by='count',ascending=False).head(5)
```

```python
# Figure Size
fig = plt.figure(figsize =(20, 10))
 
# Horizontal Bar Plot
plt.bar(n['neighbourhood_group'],n['count'])
plt.title('Attack type Vs Number of events',fontweight ='bold', fontsize = 20)
plt.xlabel('Attack type', fontweight ='bold', fontsize = 15)
plt.ylabel('Number of events', fontweight ='bold', fontsize = 15)
 
# Show Plot
plt.show()
```
